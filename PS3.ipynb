```python
import pandas as pd

users = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users_rd = pd.read_csv(users, sep='|')

#Discover what is the mean age per occupation
age_asper_occupation = users_rd.groupby('occupation')['age'].mean()
print(age_asper_occupation)

#Discover the Male ratio per occupation and sort it from the most to the least
m_ratio = users_rd.groupby('occupation')['gender'].apply(lambda x: (x == 'M').sum() / x.count())
sorted_male = m_ratio.sort_values(ascending=False)
print("\n")
print(sorted_male)

#For each occupation, calculate the minimum and maximum ages
age_max = users_rd.groupby('occupation')['age'].max()
age_min = users_rd.groupby('occupation')['age'].min()
print("\n")
print(age_max)
print("\n")
print(age_min)

#For each combination of occupation and sex, calculate the mean age
avg_age = users_rd.groupby(['occupation','gender'])['age'].mean()
print("\n")
print(avg_age)

#For each occupation present the percentage of women and men

gender_pop = users_rd.groupby(['occupation','gender'])['gender'].count()
print("\n")
print(gender_pop)


```

```python
# Step 1. Import the necessary libraries

import pandas as pd
import datetime

# Step 2. Import the dataset from the attached file wind.txt
# Step 3. Assign it to a variable called data and replace the first 3 columns by a proper
# datetime index.
wind_data = 'wind.txt'
data = pd.read_csv(wind_data, sep="\s+", parse_dates=[[0, 1, 2]])
data.head()
```

```python
# Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fix_date(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_date)

data.head()
```

```python
# Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
#date type datetime64
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

#Set index
data = data.set_index('Yr_Mo_Dy')

data.head()
```

```python
# Question 7
# Create a vertical bar chart comparing the number of marriages and divorces per
# capita in the U.S. between 1900, 1950, and 2000.
# Don't forget to label your axes!


import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
# Extract data for years 1900, 1950, and 2000
years = [1900, 1950, 2000]
selected_data = data[data['Year'].isin(years)]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot vertical bar chart for marriages per capita
ax.bar(selected_data['Year'], selected_data['Marriages_per_1000'], width=0.4, label='Marriages_per_1000')

# Plot vertical bar chart for divorces per capita
ax.bar(selected_data['Year'] + 0.4, selected_data['Divorces_per_1000'], width=0.4, label='Divorces_per_1000')

# Set labels and title
ax.set_xlabel("Year")
ax.set_ylabel("Number per Capita")
ax.set_title("Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)")

# Add legend
ax.legend()

# Set x-axis ticks
ax.set_xticks(selected_data['Year'] + 0.2)
ax.set_xticklabels(years)

# Show the plot
plt.show()
```

```python
# Question 8:
# Create a horizontal bar chart that compares the deadliest actors in Hollywood. Sort
# the actors by their kill count and label each bar with the corresponding actor's name.
# Don't forget to label your axes!

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('actor_kill_counts.csv')  # Replace 'your_data.csv' with your actual CSV file name

# Sort the data by kill count in descending order
sorted_data = data.sort_values(by='Count', ascending=False)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot horizontal bar chart
ax.barh(sorted_data['Actor'], sorted_data['Count'])

# Label each bar with the corresponding actor's name
for i, value in enumerate(sorted_data['Count']):
    ax.text(value, i, str(value), ha='left', va='center')

# Set labels and title
ax.set_xlabel("Kill Count")
ax.set_ylabel("Actor")
ax.set_title("Deadliest Actors in Hollywood")

# Show the plot
plt.show()
```

```python
#Question 10:
# Create a scatter plot showing the relationship between the total revenue earned by
# arcades and the number of Computer Science PhDs awarded in the U.S. between
# 2000 and 2009.
# Don't forget to label your axes!
# Color each dot according to its year
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('arcade-revenue-vs-cs-doctorates.csv') 


years = df['Year']
revenues = df["Total Arcade Revenue (billions)"]
phd_awards = df["Computer Science Doctorates Awarded (US)"]

# Create a figure and axis
fig, ax = plt.subplots()

for i in range(len(years)):
    ax.scatter(phd_awards[i], revenues[i], label=str(years[i]))

ax.set_xlabel("Number of Computer Science PhDs Awarded")
ax.set_ylabel("Total Revenue Earned by Arcades")
ax.set_title("Relationship between Total Revenue and CS PhDs (2000-2009)")


ax.legend()


plt.show()
```
